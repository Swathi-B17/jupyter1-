```python
from sklearn.decomposition import PCA
import sklearn as sk 
sk.preprocessing.scale
```

```python
pcs=PCA()
pca_values=pcs.fit_transform(df)
```

```python
type(pca_values)
```

```python
var=pcs.explained_variance_ratio_
var
```

```python
pccomp = pd.DataFrame(pcs.components_)
pccomp
```

```python
import numpy as np
cumvar = np.cumsum(np.round(var,decimals = 4)*100)
cumvar
```

```python
from matplotlib import pyplot as plt
plt.plot(cumvar,color="red")
plt.show()
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=0)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
wcss
```

```python
from matplotlib import pyplot as plt
plt.plot(range(1,11),wcss)
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++',random_state=0)
y_means=kmeans.fit_predict(df)
C=kmeans.cluster_centers_
```
