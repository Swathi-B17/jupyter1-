```python
import matplotlib.pyplot as plt
division=['A','B','C','D']
marks=[78,67,55,46]
avg_marks=[50,55,60,65]
width=0.2
index=np.arange(len(division))

plt.bar(index-width,marks,width=0.4,label='Marks')
plt.bar(index+width,avg_marks,width=0.4,label='Avg_Marks')
plt.legend(loc='best')
#plt.xticks(index,division)
plt.show()
```

```python
import matplotlib.pyplot as plt
division=['A','B','C','D']
marks=[78,67,55,46]
avg_marks=[50,55,60,65]
width=0.2
index=np.arange(len(division))

plt.bar(index-width,marks,width=0.4,label='Marks')
plt.bar(index+width,avg_marks,width=0.4,label='Avg_Marks')
plt.legend(loc='best')
plt.xticks(index,division)
plt.show()
```

```python
import matplotlib.pyplot as plt
division=['A','B','C','D']
marks=[78,67,55,46]
avg_marks=[50,55,60,65]
width=0.2
index=np.arange(len(division))
plt.bar(index-width,marks,width=0.4,label='Marks')
plt.bar(index+width,avg_marks,width=0.4,label='Avg_Marks')
plt.legend(loc='best')
plt.yticks(marks,avg_marks)
plt.show()
```
