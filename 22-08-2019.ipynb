### Dimension Reduction Techniques

It comes under USL(un supervised learning)
it is used when the variables are multicollinearity
We will reduce the dimensions of the dataset w.r.t columns.
There are 2 types algorithms in DRT
  1.Principal Component Analysis
  2.Singular Value Decomposition(it is mainly used in advanced analytics)
Dimension Reduction -applications:
  a.Computational performance enhanced
  b.Face recognition
  c.Image compression
PCA Benifits:
    REduce no.of columns
    Identify relation b/w columns
    Visualize multidimensional data in 2D
PCA is also called as less tranparent algorithm

### Clustering-Elbow Method

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++',random_state=0)  #k means++ method is used to find the best K initial position 
                                                                    #of the the centroids

    kmeans.fit(new_df)
    wcss.append(kmeans.inertia_)
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlim(0, 12)
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++',random_state=0)
y_kmeans=kmeans.fit_predict(new_df)
C=kmeans.cluster_centers_
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure()
ax=Axes3D(fig)
ax.scatter(new_df[0],new_df[1],new_df[2],c='y')
ax.scatter(C[0],C[1],C[2],marker='*',s=1000)
```
