```python
import pandas as pd
import numpy as np
```

```python
wc=pd.read_csv("Desktop/Womens Clothing E-Commerce Reviews.csv")
wc.head()
```

```python
wc.shape
```

```python
wc.columns
```

```python
wc.isnull().sum()
```

```python
wc.fillna("0", inplace = True)

#type(m)
```

```python
wc.Title.value_counts().sort_values(ascending=False).head(20)
```

```python
top_20=[x for x in wc.Title.value_counts().sort_values(ascending=False).head(20).index]
#top_20
```

```python
for x in top_20: 
    wc[x]=np.where(wc['Title']==x,1,0)
encoded = wc[["Title"]+top_20]
encoded
```

```python
encoded.shape
```

```python
encoded.iloc[:,1:21]
```

```python
data=pd.concat([wc,encoded],axis=1)
data.head()
#df = pd.concat([n, Y], axis=1)
# Dropping drive-wheels, make and engine-location columns as they are encoded
data.drop(['Title', 'Review Text', 'Division Name','Department Name','Class Name'], inplace=True, axis=1)
## # Viewing few rows of data
data.head()
```

```python
input_data=data.drop(data.iloc[:,[3]], axis = 1)
input_data.head(5)
```

```python
output_data=data.iloc[:,3:4]
output_data.head(5)
```

```python
from sklearn.preprocessing import scale
```

```python
input_data=scale(input_data)
input_data
```

```python
from sklearn.model_selection import train_test_split
input_data_train,input_data_test,output_data_train,output_data_test=train_test_split(input_data,output_data,test_size=0.2,random_state=0)
```

```python
input_data_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(input_data_train)
input_data_train = scaler.transform(input_data_train)
input_data_test = scaler.transform(input_data_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=2)
classifier.fit(input_data_train,output_data_train)
```

```python
output_data_pred=classifier.predict(input_data_test)
output_data_pred
```

```python
import numpy as np
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(output_data_pred,output_data_test)
```
