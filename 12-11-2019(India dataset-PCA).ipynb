### The aim of principal components analysis (PCA) is generaly to reduce the number of dimensions of a dataset where dimensions are not completly decorelated. PCA provides us with a new set of dimensions, the principal components (PC).

### Principal Component Analysis (PCA) is one of famous techniqeus for dimension reduction, feature extraction, and data visualization.

```python
df.describe()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
import sklearn as sk
```

```python
df_normal=scale(df)
type(df_normal)
```

```python
df_normal1=pd.DataFrame(df_normal)
df_normal1.head()
type(df_normal1)
```

```python
pcs=PCA()
pca_values=PCA().fit_transform(df_normal1)
pca_values=pcs.fit_transform(df_normal1)
```

```python
var = pcs.explained_variance_ratio_
var
var = pd.DataFrame(var)
var
```

```python
pccomp = pd.DataFrame(pcs.components_)
pccomp
```

```python
import numpy as np
cumvar = np.cumsum(np.round(var,decimals = 4)*100)
cumvar
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(cumvar,color="green")
plt.show()
```

### Clustering Elbow method

```python
new_df = pd.DataFrame(pca_values[:,0:5])
new_df.head()
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++',random_state=0)  #k means++ method is used to find the best K initial position                                                                 #of the the centroids
    kmeans.fit(new_df)
    wcss.append(kmeans.inertia_)
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlim(0, 12)
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```
